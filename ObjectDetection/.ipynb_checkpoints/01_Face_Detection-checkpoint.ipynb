### Import Libraries

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

### Create Classifier

```python
face_classifier = cv2.CascadeClassifier('Haarcascades/haarcascade_frontalface_default.xml')
```

### Open Image

```python
image = cv2.imread('Images/eye_face.jpg')
fix_img = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(fix_img)
```

### Detect Faces

```python
faces = face_classifier.detectMultiScale(image, 1.3, 5)
```

### No Faces Detected

```python
if faces is ():
    print("no faces found")
```

### It's a kind of Magic

```python
def detect_faces(fix_img):
    face_rects = face_classifier.detectMultiScale(fix_img, 1.3, 5)
    
    for (x,y,w,h) in face_rects:
        cv2.rectangle(fix_img, (x,y), (x+w, y+h), (255,0,0), 4)
    
    return fix_img
```

```python
plt.figure(figsize = (15,10))
result = detect_faces(fix_img)
plt.imshow(result)
```
