### Import Libraries

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

### Open Image

```python
img = cv2.imread('Images/car_plate.jpg')
```

### Convert Image

```python
def display(img):
    fig = plt.figure(figsize = (10,8))
    ax = fig.add_subplot(111)
    new_img = cv2.cvtColor(img, cv2.COLOR_RGB2BGR)
    ax.imshow(new_img)
```

### Display it

```python
display(img)
```

### Create Classifier

```python
plate_classifier = cv2.CascadeClassifier('Haarcascades/haarcascade_plate_number.xml')
```

### It's a Kind of Magic

```python
def detect_number_plate(img):
    plate_rect = plate_classifier.detectMultiScale(img)
    for (x, y, w, h) in plate_rect:
        cv2.rectangle(img, (x,y), (x+w, y+h), (5,100,10), 3)
    return img
```

### Check the Results

```python

result = detect_number_plate(img)
plt.figure(figsize = (20,18))
plt.imshow(result)
```
